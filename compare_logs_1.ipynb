```python
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
hp_866a = pd.read_csv('.\\Compare_logs\\HP-866A_anemometer_2023-02-26_15-03-26.csv')
testo405i = pd.read_csv('.\\Compare_logs\\Testo405i_2023-02-26_15-03-58.csv')
print(hp_866a.keys())
print(testo405i.keys())
```

```python
#convert string time to Unix time
t = hp_866a['time']
hp_866a['timestamp'] = hp_866a['time'].map(lambda x: time.mktime(time.strptime(x, '%d.%m.%Y %H:%M:%S')))
```

```python
plt.figure(figsize=(16,8))
t0 = min(hp_866a['timestamp'][0], testo405i['Timestamp'][0])
plt.plot(hp_866a['timestamp'] - t0, hp_866a['measurement'], label='HP-866A')
plt.plot(testo405i['Timestamp'] - t0, testo405i['Velocity m/s'], label='Testo405i')
#plt.ylim(0,5)
plt.grid()
# t = testo405i['Timestamp']
# plt.xlim(min(t), max(t))
plt.title('Скорость, м/с')
plt.legend()
plt.xlabel('t, c')
plt.ylabel('м/с')
```

```python
plt.figure(figsize=(16,8))
plt.plot(hp_866a['timestamp'] - t0, hp_866a['temperature'], label='HP-866A')
plt.plot(testo405i['Timestamp'] - t0, testo405i['Temperature C'], label='Testo405i')
# plt.ylim(0,5)
plt.grid()
#plt.xlim(min(t), max(t))
plt.title('Температура')
plt.legend()
plt.xlabel('t, c')
plt.ylabel('T, °C')
```
